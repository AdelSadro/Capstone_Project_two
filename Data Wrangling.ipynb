<font color=Maroon size=5>Combine Gas Turbine Data from 2011 to 2015<font>

```python
dataFrames = [gt_2011, gt_2012, gt_2013, gt_2014, gt_2015]
year = [2011, 2012, 2013, 2014, 2015]
for i in range(5):
    dataFrames[i]['YEAR'] = year[i]
    print(dataFrames[i].columns)
```

```python
gt = pd.concat(dataFrames, ignore_index=True)
```
